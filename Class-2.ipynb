## Random Walk

```python
'''A guy is drunk! He either walks one step ahead
or one step back each second. His each step is 1ft long.
How far will he go in 100 steps?'''
import random
def walk1D(x,N,d):
    X=[] # To store position
    T=[] # To store Time
    for t in range(N):
        x=x+d*random.choice([-1,1]) # Each walk
        T.append(t) # collect time
        X.append(x) # collect position
    return T,X  
```

```python
time,pos=walk1D(x=0,N=100,d=1)
```

```python
plt.plot(time,pos)
```
